```r
my_tubes = rep(2,12)
my_tubes
```

```r
# ml
my_amount = c(
    2.00E-02,
2.00E-03,
2.00E-04,
2.00E-05,
2.00E-06,
2.00E-07,
2.00E-08,
2.00E-09,
2.00E-10,
2.00E-11,
2.00E-12,
2.00E-13
)
my_amount
```

```r
 run_mpn <- function(sname) {
    my_positive_growth = df %>% 
        filter(Sample==sname) %>%
        arrange(column)
     print(my_positive_growth$Positive_growth)
    my_mpn <- mpn(my_positive_growth$Positive_growth, my_tubes, my_amount * 1e3, CI_method ='LR' )
    return(my_mpn)
}

```

```r
run_mpn('A0A')
```

```r
# Sample function returning a list
# cells / ul
MULTVAL =  1e3
my_function <- function(Positive_growth, n) {
    #print(group_data)
    my_mpn <- mpn(Positive_growth, n, my_amount * MULTVAL)
    return (my_mpn)
  #return(list(mean_value = mean(group_data$Positive_count), 
  #            sd_value = sd(group_data$Positive_count)))
}

# Apply the function with summarise
summary_df <- df %>%
  group_by(Sample, day) %>%
  summarise(
      MPN = my_function(Positive_growth, n)$MPN * MULTVAL, 
      MPN_adj = my_function(Positive_growth, n)$MPN_adj * MULTVAL, 
      variance = my_function(Positive_growth, n)$variance , 
      var_log = my_function(Positive_growth,n )$var_log, 
      LB = my_function(Positive_growth,n )$LB * MULTVAL, 
      UB = my_function(Positive_growth,n )$UB * MULTVAL, 
      RI = my_function(Positive_growth,n)$RI, 
  )
      #MPN = run_", "MPN_adj", "variance", "var_log", "conf_level", "CI_method", 
#"LB", "UB", "RI")
 #     mean = my_function(Positive_count)$mean_value,  # Extract values by name
 #           
            #sd = my_function(Positive_count)$sd_value)
```
